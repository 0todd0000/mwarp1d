**Figure 3**. (a) Manual warp parameters: `center`, `amp` (amplitude), `head` and `tail`. Data at position `center` will be translated horizontally with an approximate magnitude of `amp`, where `head` and `tail` determine translation effect fall-off, or the translation effects on neighboring regions. (b) Three example warp kernels (i.e., displacement fields). (c) Corresponding warped domains. (d) Corresponding warped data for a single 1D measurement from Fig.1.

### Summary

`mwarp1d` allows users to nonlinearly register 1D data using either (1) landmark warping or (2) manual warping. This can be achieved via scripting, as demonstrated above, or using the `mwarp1d` GUI, as described in a series of [screencasts](http://www.spm1d.org/mwarp1d/quickstart.html).
